```python
import json
import pandas as pd
```

Predictions is the result from Josh's frame.

```python
with open("./data/predictions.json", "r") as json_file:
    predictions = json.load(json_file)

# Create a DataFrame from the JSON data
betters = pd.DataFrame(predictions)

# explode the eth column, so that each element in the list is a new row
betters = betters.explode("eth")

betters.head()
```

erc20 transfers is from basescan (maybe a cleverer way to get at it), and represents everyone who gave a erc20 token to the football party contract from March 25th (arbitrary) to March 31st.

```python
erc20_transfers = pd.read_csv("./data/erc20_transfers.csv")
erc20_transfers.head()
```

Merge the two to find out who made a prediction, and how much they contributed

```python
df = pd.merge(erc20_transfers, betters, left_on="From", right_on="eth")
df.head()
```

```python
results = df[["username", "TokenName", "TokenSymbol", "TokenValue"]]
results.head(10)
```
